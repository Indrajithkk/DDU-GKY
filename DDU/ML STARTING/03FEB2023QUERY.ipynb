## ASSIGN

```python
##CREATING A NEW COLUMN WITH ASSIGN FUNCTION-
```

```python
import pandas as pd
import numpy as  np 
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('netflix_titles_2021.csv')
df.head(5)
```

```python
## create a column in [df.assign] function. this function will create a new column this value will be 'true' or ' false'
df.assign(after_2020=df["release_year"]>2020)
```

```python
##after 2000 movies column
df.assign(after_2000=df["release_year"]<2000)
```

## QUERY

```python
## "df.query" This function help to find the certain raws or columns
## find the 2020 released movies
df.query('release_year==2020 & type=="Movie"')
```

```python
##find the 2020 relesed tv shows
df.query('release_year==2020&type=="TV Show"')
```

```python
df.query('release_year==2020&show_id=="s18"')
```
