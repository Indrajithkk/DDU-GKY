```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("Car_sales.csv")
```

```python
data
```

## droping na values 

```python
data.dropna(inplace=True)
```

```python
data.info()
```

## plots on numerical values

```python
### ploting
data.Horsepower.plot.hist(figsize=(6,3), cmap= 'brg') ;
```

```python
## create box plot
data.Sales_in_thousands.plot.box(figsize=(6,6), cmap= 'rainbow');
```

```python
data.Horsepower.plot.box(figsize=(6,3), cmap= 'Wistia_r') ;
```

## sub plots

```python
data.plot.box(figsize=(22,3), subplots= True);
```

data.plot.hist(figsize=(6,22), subplot = True );

```python
data.plot.hist(figsize=(3,20), subplots = True );
```

```python
data.sort_values(by='__year_resale_value').plot.line('__year_resale_value','Price_in_thousands',figsize=(6,3),cmap='Wistia_r');
```

## bar plot

```python
data.Manufacturer.value_counts().plot.bar(figsize=(6,2),cmap='twilight_shifted')
```

## Area chart

```python
data.plot.area('Manufacturer','Power_perf_factor', figsize=(10,3),cmap='gist_rainbow');
```

## Horizontal bar chart

```python
data.sort_values( by ='Price_in_thousands')[:5].plot.barh('Model','Power_perf_factor', figsize=(6,4));
```

## scatter plot

```python
data.sort_values( by ='Price_in_thousands')[:5].plot.scatter('Model','Power_perf_factor', figsize=(6,4));
```
