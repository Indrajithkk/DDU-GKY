## Outliear treatment

### BOXPLOT

```python
## A for loop( for i in) is used for iterating over a sequence.

for i in data.columns:
    if data[i].dtype != "object": ## if the data is 'i' datatype not equal to object because this used for only numerical values
        sns.boxplot(y=data[i], color="yellow") ## creating seaborn boxplot. y= another variable to assign the value of whitout object datatype
        plt.show(); ## plt means plot. and show() is print the graph
        
```

## SETTING OUTLIEAR LIMITS:

```python
def outliear_limit(col):
    Q3,Q1=np.nanpercentile(col,[75,25])
    IQR = Q3-Q1
    UL = Q3+1.5*IQR
    LL = Q1-1.5*Q1
    return UL,LL
```

```python
for column in data.columns:
    if data [column].dtype !="object":
        UL,LL = outliear_limit(data[column])
        data[column] = np.where((data[column]>UL)|(data[column]<LL), np.nan, data[column])
```

#### LETS PLOT THE BARPLOT AGAIN TO SEE DIFFERENCE

```python
for i in data.columns:
    if data[i].dtype != "object": ## if the data is 'i' datatype not equal to object because this used for only numerical values
        sns.boxplot(y=data[i], color="red") ## creating seaborn boxplot. y= another variable to assign the value of whitout object datatype
        plt.show();
```

### We can see the outliers are treated
